## Tweet collection for topic modelling
___

* Received a csv file from Alan ( The list of users )
* It contains : 34,872  unique users
* Tweets are collected between dates 2014-09-01 to 2015-02-26

####  Time required 
___

Collection takes aroung 25-30 mins for every 100 users




```python
import pandas as pd
from pymongo import MongoClient
```

```python
conn = MongoClient("localhost", 27017)
db = conn['tweet_db']

print (db)
```

```python
collection_name = "AmericanCrime_tweets"
collection = db[collection_name]
```

```python
count = collection.count()
print(count)
```

```python
cursor = collection.find()
df = pd.DataFrame(list(cursor))
print(df[:20])
```

```python
df.groupby('screen_name').count()
```

```python
# users with no tweets
collection_scren_name = "AmericanCrime"
collection_s = db[collection_scren_name]
count_screenname = collection_s.count()

print(34872 - count_screenname)
```
